```python
import numpy as np
import pandas as pd
import time as time
import matplotlib.pyplot as plt
```

```python
df_1k = pd.read_csv('results_1k.csv')
df_10k = pd.read_csv('results_10k.csv')
df_100k = pd.read_csv('results_100k.csv')
df_1m = pd.read_csv('results_1m.csv')
df_2m = pd.read_csv('results_2m.csv')
df_3m = pd.read_csv('results_3m.csv')
df_4m = pd.read_csv('results_4m.csv')
df_5m = pd.read_csv('results_5m.csv')
df_10m = pd.read_csv('results_10m.csv')
df_fixed_s = pd.read_csv('results_fixed_s.csv')
```

```python
def time_complexity(n, s=10):
    return n * np.log(n / s) + n * s

# Array sizes to compare
n_values = np.arange(1000, 10000000,1000)  # n from 1,000 to 5,000,000 with step size of 10,000

# Fixed value of s = 10
s = 10

# Calculating the time complexity for different values of n
time_complexities = [time_complexity(n, s) for n in n_values]

# Plotting the time complexity
plt.figure(figsize=(10, 6))
plt.plot(n_values, time_complexities, label='theoretical', color='red')


# With the value of S fixed, plot the number of key comparisons over different sizes of the input list n. 

plt.title("Key Comparisons Over Different Array Sizes") 
plt.xlabel("Array Size") 
plt.ylabel("Key Comparisons")
plt.ticklabel_format(axis="y", style="plain")
plt.ticklabel_format(axis="x", style="plain")

#for xy in zip(df_fixed_s.dataset_size, df_fixed_s.keyComp): #show points
   #plt.annotate('(%d, %d)' % xy, xy=xy)

plt.plot(df_fixed_s.dataset_size, df_fixed_s.keyComp, marker = 'o', label = 'empirical')

plt.legend()
```
