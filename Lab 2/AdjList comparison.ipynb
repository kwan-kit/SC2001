```python
import numpy as np
import pandas as pd
import time as time
import matplotlib.pyplot as plt
```

# Part A

```python
df_fixedV = pd.read_csv('SC2001_Project2_Djikstra\AdjMatrix_FixedV_Results.csv')
df_fixedE = pd.read_csv('SC2001_Project2_Djikstra\AdjMatrix_FixedE_Results.csv')
```

```python
def time_complexity(v, e):
    return v*v

# Array sizes to compare
i_values = np.arange(1000, 9900, 100)  # i from 1,000 to 9900 with step size of 100

# Calculating the time complexity for different values of V and E
fixedV_time_complexities = [time_complexity(100, i) for i in i_values]
fixedE_time_complexities = [time_complexity(i, 20000) for i in i_values]
```

```python
# Plotting the time complexity (Fixed V, Varying E)
plt.figure(figsize=(10, 6))
plt.plot(i_values, fixedV_time_complexities, label='theoretical', color='red')

plt.title("Key Comparisons Over Different Number of Edges") 
plt.xlabel("Edges") 
plt.ylabel("Key Comparisons")
plt.ticklabel_format(axis="y", style="plain")
plt.ticklabel_format(axis="x", style="plain")

plt.plot(df_fixedV.Edges, df_fixedV.KeyComp, marker = 'o', label = 'empirical')

plt.legend()
```

```python
# Plotting the time complexity (Varying V, Fixed E)
plt.figure(figsize=(10, 6))
plt.plot(i_values, fixedE_time_complexities, label='theoretical', color='orange')

plt.title("Key Comparisons Over Different Number of Vertices") 
plt.xlabel("Vertices") 
plt.ylabel("Key Comparisons")
plt.ticklabel_format(axis="y", style="plain")
plt.ticklabel_format(axis="x", style="plain")

plt.plot(df_fixedE.Vertices, df_fixedE.KeyComp, label = 'empirical', linestyle='--', color = 'blue')

plt.legend()
```

```python
# Plotting the time complexity (Varying V, Fixed E)
plt.figure(figsize=(10, 6))
plt.plot(i_values, fixedE_time_complexities, label='theoretical', color='red')

plt.title("Key Comparisons Over Different Number of Vertices") 
plt.xlabel("Vertices") 
plt.ylabel("Key Comparisons")
plt.ticklabel_format(axis="y", style="plain")
plt.ticklabel_format(axis="x", style="plain")

plt.plot(df_fixedE.Vertices, df_fixedE.KeyComp, marker = 'o', label = 'empirical')

plt.legend()
```

```python
# Plotting the runtime (Fixed V, Varying E)
plt.figure(figsize=(10, 6))

plt.title("Runtime Over Different Number of Edges") 
plt.xlabel("Edges") 
plt.ylabel("Runtime (ms)")
plt.ticklabel_format(axis="y", style="plain")
plt.ticklabel_format(axis="x", style="plain")

plt.plot(df_fixedV.Edges, df_fixedV.ElapsedTime, marker = 'o')
```

```python
# Plotting the runtime (Varying V, Fixed E)
plt.figure(figsize=(10, 6))

plt.title("Runtime Over Different Number of Vertices") 
plt.xlabel("Vertices") 
plt.ylabel("Runtime (ms)")
plt.ticklabel_format(axis="y", style="plain")
plt.ticklabel_format(axis="x", style="plain")

plt.plot(df_fixedE.Vertices, df_fixedE.ElapsedTime, marker = 'o')
```
